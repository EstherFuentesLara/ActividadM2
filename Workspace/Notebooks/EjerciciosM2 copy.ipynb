```python
# Funcion para visualizar relacion de una variable con el objetivo
def visualizar_relacion(variable):
    
    plt.figure(figsize=(6,4))
    sns.boxplot(x="TenYearCHD", y=variable, data=dataframe1)
    
    plt.title(f"{variable} vs riesgo cardiovascular")
    plt.show()
```

```python
# Relación edad - enfermedad
# Los pacientes con enfermedad cardiovascular presentan una mediana de edad mayor,
# lo que sugiere una posible relación entre edad avanzada y mayor riesgo cardiovascular.
visualizar_relacion("age")
```

```python
def visualizar_relaciones(variables):
    
    fig, axes = plt.subplots(1, len(variables), figsize=(20,4))
    
    for i, var in enumerate(variables):
        sns.boxplot(x="TenYearCHD", y=var, data=dataframe1, ax=axes[i])
        axes[i].set_title(f"{var} vs riesgo cardiovascular")
    
    plt.tight_layout()
    plt.show()
```

```python
# Para una mejor visualizacion unificamos en una unica imagen todas las varibles
visualizar_relaciones(["age","sysBP","diaBP","glucose","BMI"])
```

Si lo estimas necesario, ejecuta métodos de imputación y normalización de datos explicando explicando brevemente tus decisiones.

```python
# Comprobamos si tiene valores nulos
x=dataframe2.isnull().sum()
x[x!=0]
```

```python
# Se eliminan los valores nulos
dataframe3=dataframe2.dropna()
dataframe3.isnull().sum()
```

Realizamos tratamiento de valores atipicos. 
En este contexto, los valores atípicos no se eliminan ya que pueden representar
casos clínicos reales (por ejemplo, pacientes con hipertensión severa o niveles
elevados de glucosa). En el ámbito sanitario, estos valores son especialmente
relevantes porque están asociados a un mayor riesgo de enfermedad.
Eliminarlos podría implicar la pérdida de información importante.
En cualquier caso, si aplicaremos un escalado robusto en la etapa de normalización de los datos. 
Esto reduce el impacto de valores atipicos sin necesidad de elimiarlos.

```python
# Quitamos duplicados
dataframe4=dataframe3.drop_duplicates()
```

 El escalado se aplica únicamente a variables continuas debido a sus diferentes
 escalas y unidades de medida. Las variables binarias se mantienen sin transformar,
 ya que ya están codificadas en formato 0/1 y su normalización podría afectar a su interpretación.

```python
# Normalizamos las variables acorde al modelo que vamos a elegir (en este caso Random Forest)
from sklearn.preprocessing import RobustScaler

# -- Separación de variables 
# Columnas que SÍ se escalan (continuas)
columnas_continuas = [
    "age", "cigsPerDay", "totChol", "sysBP", "diaBP", "BMI", "heartRate", "glucose"
]
# Columnas que NO se escalan (binarias + education)
columnas_no_escalar = [
    "male", "BPMeds", "prevalentStroke", "diabetes", "education","TenYearCHD"
]

# -- Separar datasets
df_continuas = dataframe4[columnas_continuas]
df_no_escalar = dataframe4[columnas_no_escalar]

# -- Normalización (escalado robusto)
scaler = RobustScaler()

df_continuas_scaled = pd.DataFrame(
    scaler.fit_transform(df_continuas),
    columns=columnas_continuas,
    index=dataframe4.index
)

# -- Añadir variables no escaladas
df_scaled = pd.concat([df_continuas_scaled, df_no_escalar], axis=1)
```

# Ejercicio 5.
Partiendo de esta submuestra final, genera los grupos de entrenamiento y validación e imprime pos pantalla alguna características del mismo.

```python
import numpy as np
from sklearn.model_selection import KFold
```

```python
# Vamos a definir X e Y
# X -> entradas
# Y -> salidas esperadas
X=df_scaled.drop(["TenYearCHD"],axis=1)
y=df_scaled['TenYearCHD']
```

Realizamos la separacion con Kfold
    n_splits --> es el numero de partes que vamos a hacer que coincide con el numero de pruebas que voy a hacer
    random_state --> para que siempre sea la misma prueba numero aleatorio

Se emplea validación cruzada K-Fold con k=5 para obtener una evaluación
más robusta del modelo. Este método permite que todos los datos participen
tanto en entrenamiento como en validación, reduciendo la variabilidad asociada
a una única partición. Se selecciona k=5 como equilibrio entre coste computacional
y estabilidad en la estimación del rendimiento.

```python
# n_splits --> es el numero de partes que vamos a hacer que coincide con el numero de pruebas que voy a hacer
# random_state --> para que siempre sea la misma prueba numero aleatorio

kf=KFold(n_splits=5, random_state=37, shuffle=True)
kf.split(X,y)
```

```python
X_train, X_test, y_train, y_test = ([], [], [], [])

for i, (train_index, test_index) in enumerate(kf.split(X)):
   X_train.append(X.iloc[train_index, :])
   X_test.append(X.iloc[test_index, :])
   y_train.append(y.iloc[train_index, ])
   y_test.append(y.iloc[test_index, ])
```

# Ejercicio 6.
 Instancia el modelo Scikit-learn de tu elección y aplícalo sobre los datos seleccionados para el entrenamiento.

 Se utiliza Random Forest por su capacidad para modelar relaciones no lineales,
 su robustez frente a outliers y su buen rendimiento en problemas de clasificación
 con variables heterogéneas como en el dataset Framingham.

 En consecuencia, Random Forest representa un buen equilibrio entre rendimiento, robustez y facilidad de uso
 comparado con otros modelos.

 Se ha aplicado una configuracion de hiperparametros que limita la capacidad de aprendizaje del modelo para 
 reducir la posibilidad de sobrejuste. Se utiliza GridSearchCV para optimizar los hiperparámetros del modelo, evaluando múltiples combinaciones mediante validación cruzada.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# --- Modelo base ---
rf = RandomForestClassifier(random_state=42)

# --- Hiperparámetros a probar ---
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [5, 10, None],
    "min_samples_split": [2, 10, 20],
    "min_samples_leaf": [1, 5, 10]
}

# --- Grid Search ---
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

# --- Entrenamiento ---
grid_search.fit(X, y)

# --- Mejores parámetros ---
print("Mejores parámetros:", grid_search.best_params_)

# --- Mejor resultado ---
print("Mejor accuracy:", grid_search.best_score_)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

kf_model = []

for i in range(len(X_train)):
# Modelo 
    model = RandomForestClassifier(
        n_estimators=100, 
        max_depth=10,
        min_samples_leaf=10,
        min_samples_split=2,
        random_state=42
    )

    model.fit(X_train[i], y_train[i])
    kf_model.append(model)
    
    y_pred = model.predict(X_train[i])
    accuracy = accuracy_score(y_train[i], y_pred)
    print("Train Accuracy:", accuracy)
```

Los resultados obtenidos en la prediccion de test es muy similar a la del entrenamiento indicando que el modelo

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

for i in range(len(X_test)):
    # --- Predicción ---
    y_pred = kf_model[i].predict(X_test[i])

    # --- Accuracy ---
    accuracy = accuracy_score(y_test[i], y_pred)
    print("Accuracy:", accuracy)

    # --- Reporte completo ---
    print("\nReporte de clasificación:")
    print(classification_report(y_test[i], y_pred))

    # --- Matriz de confusión ---
    print("\nMatriz de confusión:")
    print(confusion_matrix(y_test[i], y_pred))
```
